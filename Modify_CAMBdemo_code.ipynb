```python
# massively pared-down version of the code at https://camb.readthedocs.io/en/latest/CAMBdemo.html
# trimmed down to a MWE of the MPS
from matplotlib import pyplot as plt
import numpy as np
import camb
from camb import model
```

```python
pars=camb.set_params(H0=67.5, ombh2=0.022, omch2=0.122, ns=0.965)
pars.set_matter_power(redshifts=[3.0,5.0,10.0], kmax=2.0)
lin=True
results = camb.get_results(pars)
if lin:
    pars.NonLinear = model.NonLinear_none
else:
    pars.NonLinear = model.NonLinear_both
    
kh,z,pk=results.get_matter_power_spectrum(minkh=1e-4,maxkh=1,npoints=200)
```

```python
plt.figure()
plt.loglog(kh,pk[0,:],label=str(z))
plt.xlabel('k/h Mpc')
plt.ylabel('power')
plt.legend()
plt.title('Matter power at z='+str(z))
plt.show()
```
